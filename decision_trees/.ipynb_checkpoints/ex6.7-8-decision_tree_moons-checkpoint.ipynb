```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split, GridSearchCV, ShuffleSplit
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Generate moons dataset
X, y = make_moons(n_samples=10000, noise=0.4, random_state=42)
```

```python
def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
```

```python
# Split into traing and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
plot_dataset(X_train, y_train, [-1.5, 2.5, -1, 1.5])
plt.show()
```

```python
# Use grid search with cross-validation to find good hyperparameters
params = {
    "max_leaf_nodes": [None, 10, 20, 30, 40],
    "max_depth": [1, 5, 10, 20, 40],
}

grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42), param_grid=params, cv=3
)
```

```python
grid_search.fit(X_train, y=y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
y_pred = grid_search.best_estimator_.predict(X_test)
accuracy_score(y_test, y_pred)
```

# Ex6.8: Grow a forrest

```python
# Gerenerate 1,000 subsets of the training data. Each containing 100 randomly selected instances. Use ShuffleSplit class
shuffled_samples = ShuffleSplit(n_splits=1000, train_size=100, random_state=42)

subsets = list()
for train_indicies, _ in shuffled_samples.split(X_train):
    X_train_subset = X_train[train_indicies]
    y_train_subset = y_train[train_indicies]
    subsets.append((X_train_subset, y_train_subset))
```

```python
# Train one dt_clf on each subset using best model found in 6.9
from sklearn.base import clone

forest = [clone(grid_search.best_estimator_) for _ in range(1000)]

accuracy_scores = list()

for tree, (X_mini_train, y_mini_train) in zip(forest, subsets):
    tree.fit(X_mini_train, y_mini_train)
    
    y_pred = tree.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    
np.mean(accuracy_scores)
```

```python
# For each test set instance generate the predictions of the 1k decisions trees and keep only the most frequent prediction
from scipy import stats

y_pred = list()
for X_instance, y_instance in zip(X_test, y_test):

    predictions = list()
    for tree in forest:
        predictions.append(tree.predict([X_instance]))
        
    y_pred.append(stats.mode(predictions).mode[0])
```

```python
accuracy_score(y_test, y_pred)
```
